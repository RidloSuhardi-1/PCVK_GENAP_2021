```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as im
```

```python
def convolution2d(image, kernel, stride, padding):
  img = np.pad(image, padding, mode='constant', constant_values=0)

  kernel_height, kernel_width = kernel.shape
  pad_height, pad_width = img.shape

  output_width = (pad_width - kernel_width) // stride + 1
  output_height = (pad_height - kernel_height) // stride + 1

  foo_img = np.zeros((output_height, output_width)).astype(np.float32)

  for y in range(0, output_height):
    for x in range(0, output_width):
      foo_img[y][x] = np.sum(image[y * stride:y * stride + kernel_height, x * stride:x * stride + kernel_width] * kernel).astype(np.float32)
  return foo_img
```

```python
get_img = cv.imread('/content/drive/MyDrive/Images/Images/mandrill.tiff')
img_gray = cv.cvtColor(get_img, cv.COLOR_BGR2GRAY)
```

```python
# image sharpen
kernel_sharpen = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])

cv2_imshow(img_gray)
cv2_imshow(convolution2d(img_gray,kernel_sharpen,1,2))

# final_frame = cv.hconcat((get_img, convolution2d(img_gray,kernel_sharpen,1,2)))
# cv2_imshow(final_frame)
```
